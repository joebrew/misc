```python
def get_max_record(group):
    return group.ix[group.prop.idxmax()]
```

```python
# apply function to whole dataset
get_max_record(boys)
```

```python
# apply function to each year.
result = boys.groupby('year').apply(get_max_record)
```

```python
# plot the result
result.prop.plot()
```

```python
def get_quantile_count(group, quantile=0.5):
    df = group.sort_index(by='prop', ascending=False)
    return df.prop.cumsum().searchsorted(quantile)
```

```python
births = read_csv('PandasTutorialFiles/births.csv')
```

```python
# JOIN / MERGE
merge?
# left, right, how= 'inner', on=, left_on, right_on, etc...
```

```python
# Join on year and sex from both dataframes
merged = merge(names, births, on =['year', 'sex']) # if you don't specify merge keys, it'll use intersection of columns
```

```python
merged['persons'] = np.floor(merged.prop * merged.births)
```

```python
merged.head()
```

```python
merged.groupby(['name', 'sex'])['persons'].sum()
```

```python
merged.groupby(['name', 'sex'])['persons'].sum().order()
```

```python
mboys = merge(boys, births)
mboys['persons'] = np.floor(mboys.prop * mboys.births)
persons = mboys.set_index(['year', 'name']).persons
```

```python
persons.ix[:,'Christopher'].plot(kind='bar', rot=90)
```
